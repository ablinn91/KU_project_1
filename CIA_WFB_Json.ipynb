```python
import csv
import matplotlib.pyplot as plt
import requests
import pandas as pd
import json
import os

# Load JSON
filepath = os.path.join("Data", "CIA_WFB_demograhics.json")
with open(filepath) as jsonfile:
    CIA_WFB_demograhics = json.load(jsonfile)
```

```python
CIA_WFB_demograhics.keys()
CIA_WFB_demograhics["countries"].keys()
```

```python
country_list = []
size_list = []
climate_list = []
pop_list = []
median_age_list = []
life_exp_list = []
roadways_paved_list = []
roadways_unpaved_list = []

countries = ["ethiopia", "japan", "kenya", "uganda", "united_states", "united_kingdom"]


for country in countries:
        
    data = CIA_WFB_demograhics["countries"][country]
    
    try:   
        country_list.append((data['data']['name']))
        size_list.append((data['data']['geography']['area']['comparative']))
        climate_list.append((data['data']['geography']['climate']))
        pop_list.append((data['data']['people']['population']['total']))
        median_age_list.append((data['data']['people']['median_age']['total']['value']))
        life_exp_list.append((data['data']['people']['life_expectancy_at_birth']['total_population']['value']))
        roadways_paved_list.append((data['data']['transportation']['roadways']['paved']['value']))
        if 'unpaved' in data['data']['transportation']['roadways'].keys():
            roadways_unpaved_list.append((data['data']['transportation']['roadways']['unpaved']['value']))
        else:
            roadways_unpaved_list.append(1)
    
    except(KeyError):
        print("Missing field/result... skipping.")
   
# CIA_WFB_demograhics["countries"]['world']['data']['transportation']['roadways']['unpaved'].keys()

```

```python
country_info_df = pd.DataFrame({'country':country_list, 
                                'size':size_list, 
                                'climate':climate_list, 
                                'population':pop_list, 
                                'median_age':median_age_list, 
                                'life_expectancy':life_exp_list, 
                                'paved_roads':roadways_paved_list, 
                                'unpaved_roads':roadways_unpaved_list
    
})
country_info_df
```
